#### pandas는 csv, xlsx 등의 데이터를 DataFrame (데이터형식)로 변환해 분석후 다시 원하는 데이터형식으로 보여줌

###  pandas를 이용하여 DataFrame 만들기

```python
# pandas 블러오기
import pandas as pd
```

```python
# 행방향으로 DataFrame 만들기 : List ex) a = [1,2,3] -> a[0] = collection
list01  = [
    [10,20,30],
    [40,40,50],
    [70,80,90],
]
df = pd.DataFrame(list01)  # 밑에서 작업하려면 변수 df 에 넣어 메모리에 저장
```

### 데이터 프레임에서 값 추출하기

```python
# 열(column) 기준으로 추출하기 : 0번 열의 데이터 추출하기 (slicing -  start : stop+1  ) . panas의 출력은 인덱스(0,1,2)가 보임 = series
df.loc[:,0]
```

```python
# 열(column) 기준으로 추출하기 : 0,1번 열의 데이터 추출하기
df.loc[:,[0,1]] # index를 사용한 추출
df.loc[:,0:1] # slicing을 사용한 추출
```

```python
# 범위를 정해 행(row) 기준으로 추출하기 
df.loc[0,:]
df.loc[0:1.:]
```

```python
# 특정 위치의 값만 추출하기
print(df.loc[1,2])
print(df.loc[1:2,1][2])
```

### Series
    - pandas에서 사용하는 기본 데이터
    - 데어터프레임을 정의할때는 1차원 리스트로 전달
    - index 자동생성

```python
# series

list01 = [1,2,3]
s1 = pd.Series(list01) # series는 1차원 리스트
print(s1[1]) # 1차원데이터이므로 열과 행이 없고 안에 ,나 []  없음
```

### Dictionary{key:value} 로 DataFrame만들기

- Data는 Scalar (a=1), vector ( a = 1,2,3) 타입이 있다


```python
dic01 ={
    "weight" :  [ '80', 70.4, 65.5, 45.9,51.2],
    "height"  : [170, 180, 155, 143,154],
    "type" : ['f','n','n','t','t']
}
tb1 = pd.DataFrame(dic01)
```

```python
tb1.info()
# weight 처음에 '80'으로 표기해서 weight는 모든 데이터가 string으로 인식하게됨. 정수와 실수를 함께 넣으면 정수도 실수로 표시된다. info로 Dtype을 볼것 (object=String)
```

```python
# weight 열만 추출
tb1.loc[:,'weight']  #  열이 문자가 들어갔으므로 문자로 바뀌어서 '이름'으로 불러와야함 index가 아닌 string 예를 들어 0 1 2 ㄴsum 이렇게 column이 되면 0 1 2도 문자로 인식 '0' 이렇게 해야
```
